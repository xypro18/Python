### Applying the DRY principle

If the probability of having to parse other csv files in the future is very high, why not create a function that saves us the initial work? Let's solve the exercise again, applying this:

```python
# we created a function that will read any csv file (we pass the file as the first argument)
# the second argument is the number of lines we want to skip
# the third argument is the name of the function that contains the logic of what we want to 
# with the data in the csv
def readCSV(file, skippers, job):
    import csv
    with open(file, 'r') as data_file:
        csv_data = csv.reader(data_file)
        if(skippers > 0):
            while skippers:
                next(csv_data)
                skippers -= 1
        job(csv_data)

def get_patrons_list(data):
    html_output = ''
    names = [] 
    for line in data:
        if line[0] == 'No Reward':
            break
        names.append(f'{line[0]} {line[1]}')

    html_output += f'<p>There are {len(names)} contributors:</p>'

    html_output += '\n<ul>'

    for name in names:
        html_output += f'\n\t<li>{name}</li>'

    html_output += '\n</ul>'

    print(html_output)
   
readCSV('patrons.csv', 2, get_patrons_list)
```
